## Словарь признаков для текущих CSV

Ниже приведена рабочая расшифровка признаков, которые есть в текущих файлах `data/gamma.csv` и `data/proton.csv` и соотносятся с PDF `Parameters_Definitions_Report.pdf`.

Важное ограничение: PDF описывает более полные экспериментальные таблицы TAIGA-IACT02, а текущие CSV являются сокращёнными таблицами признаков для задачи `gamma` vs `proton`. Поэтому информация из PDF подходит к нашим данным только частично. Хорошо применимы разделы про Hillas-параметры, моменты формы изображения и параметры клининга. В текущих CSV нет многих полей наблюдений, времени, pointing-файлов, фона и темпа счёта: `por`, `time`, `unix_time_ns`, `delta_time`, `error_deg`, `tel_az`, `tel_el`, `source_az`, `source_el`, `CR400phe`, `CR_portion`, `median_background`, `MAD_background`, `tel_ra`, `tel_dec`, `source_ra`, `source_dec`, `tracking`, `good`, `star`, `weather_mark`, `alpha_c`. Поэтому в этой работе используйте только фактически присутствующие колонки.

| `kurtosis_l` | Эксцесс вдоль главной оси. Есть в `proton.csv`. |
| `kurtosis_w` | Эксцесс вдоль малой оси. Есть в `proton.csv`. |
| `kurtosis` | Эксцесс в `gamma.csv`; при сравнении с proton-данными нужно отдельно проверить, чему он соответствует. |
| `alpha[0]` | Угол между главной осью эллипса и направлением на источник. В PDF: `alpha1`. |
| `alpha[1]` | Угол между главной осью эллипса и направлением на антиисточник. В PDF: `alpha2`. |
| `source_x` | X-координата источника в камере в градусах. |
| `source_y` | Y-координата источника в камере в градусах. |

Колонки без надежного описания из предоставленного словаря, но присутствующие в данных: `clean_number`, `axis_scatter`, `energy`, `azwidth[1]`, `miss[1]`, `x_max_coord`, `y_max_coord`, `x_ground`, `y_ground`, `tel_tet`, `tel_fi`, `source_tet`, `source_fi`, `Xmax`, `con1`, `con3`.

## Задание 2. Сравнить наборы колонок

Перед объединением таблиц нужно понять, какие колонки общие, а какие есть только в одном классе.

Для обучения классификатора обычно можно использовать только признаки, которые доступны для обоих классов.

```python
gamma_columns = set(gamma.columns)
proton_columns = set(proton.columns)

common_columns = [col for col in gamma.columns if col in proton_columns]
gamma_only = sorted(gamma_columns - proton_columns)
proton_only = sorted(proton_columns - gamma_columns)

print("Common columns:", len(common_columns))
print(common_columns)

print("\nOnly in gamma:", len(gamma_only))
print(gamma_only)

print("\nOnly in proton:", len(proton_only))
print(proton_only)
```

## Задание 3. Добавить целевой класс и объединить таблицы

Создайте целевую переменную:

- `target = 0` для `gamma`;
- `target = 1` для `proton`.

Также добавьте текстовую колонку `particle`, чтобы удобнее строить графики.

```python
gamma_work = gamma[common_columns].copy()
proton_work = proton[common_columns].copy()

gamma_work["target"] = 0
gamma_work["particle"] = "gamma"

proton_work["target"] = 1
proton_work["particle"] = "proton"

df = pd.concat([gamma_work, proton_work], axis=0, ignore_index=True)

print("Combined shape:", df.shape)
display(df.head())
display(df.tail())
```

## Задание 7. Исследовать корреляцию признаков с таргетом

Так как `target` закодирован числами `0` и `1`, обычная корреляция Пирсона между числовым признаком и `target` показывает линейную связь признака с классом.

Интерпретация:

- положительная корреляция: большие значения признака чаще соответствуют `proton`;
- отрицательная корреляция: большие значения признака чаще соответствуют `gamma`;
- корреляция около нуля не означает, что признак бесполезен: связь может быть нелинейной.

```python
corr_with_target = (
    df[numeric_columns]
    .corrwith(df["target"])
    .dropna()
    .to_frame("corr_with_target")
)
corr_with_target["abs_corr"] = corr_with_target["corr_with_target"].abs()
corr_with_target = corr_with_target.sort_values("abs_corr", ascending=False)

display(corr_with_target)
```

```python
top_n = 20
corr_top = corr_with_target.head(top_n).sort_values("corr_with_target")

plt.figure(figsize=(9, 0.45 * len(corr_top) + 2))
sns.barplot(
    data=corr_top.reset_index().rename(columns={"index": "feature"}),
    x="corr_with_target",
    y="feature",
    color="steelblue",
)
plt.axvline(0, color="black", linewidth=1)
plt.title("Correlation with target: 0 = gamma, 1 = proton")
plt.xlabel("Pearson correlation")
plt.ylabel("Feature")
plt.tight_layout()
plt.show()
```
